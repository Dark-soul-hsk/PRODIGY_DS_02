```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io

uploaded = files.upload()

if 'train.csv' in uploaded:
    TITANIC_FILE = io.BytesIO(uploaded['train.csv'])
    train_df = pd.read_csv(TITANIC_FILE)
    print("Successfully loaded 'train.csv'.")



survival_pivot = (
    train_df.groupby(['Pclass', 'Sex'])['Survived'].mean()
    .unstack()
    * 100
).round(1)

plt.figure(figsize=(7, 5))

sns.heatmap(
    survival_pivot,
    annot=True,
    fmt=".1f",
    cmap="YlGnBu",
    linewidths=.5,
)

plt.title('Survival Rate (%) by Pclass and Sex')
plt.show()
```
